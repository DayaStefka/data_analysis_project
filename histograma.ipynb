```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos (ejemplo desde CSV exportado de BigQuery)
df = pd.read_csv("datos_music.csv")

# Histograma de streams
plt.figure(figsize=(8,5))
sns.histplot(df["streams"], bins=30, kde=False)
plt.title("Distribución de Streams")
plt.xlabel("Streams")
plt.ylabel("Frecuencia")
plt.show()

# Boxplot de streams
plt.figure(figsize=(6,4))
sns.boxplot(x=df["streams"])
plt.title("Boxplot de Streams")
plt.show()
```

```python
# Conteo de canciones según número de artistas
plt.figure(figsize=(6,4))
df["artist_count"].value_counts().sort_index().plot(kind="bar")
plt.title("Distribución de canciones según número de artistas")
plt.xlabel("Número de artistas en la canción")
plt.ylabel("Cantidad de canciones")
plt.show()
```
